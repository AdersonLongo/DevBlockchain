**IMPORTAR BIBLIOTECAS**


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
# Carregar a base de dados IRIS
iris = load_iris()
```

```python
# Criar um DataFrame a partir dos dados do Iris
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(df.head())
```

```python
# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
# Plotar gráfico de dispersão para a base de treinamento
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap)
plt.title("Base de Treinamento")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

```python
# Plotar gráfico de dispersão para a base de teste
plt.figure()
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cmap)
plt.title("Base de Teste")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()
```

```python
# Criar o classificador kNN com k = 3
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinar o modelo com os dados de treinamento
knn.fit(X_train, y_train)
```

```python
# Fazer previsões no conjunto de teste
y_pred = knn.predict(X_test)
```

```python
# Calcular a acurácia das previsões
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia:", accuracy)
```

```python
# Fazer previsões no conjunto de teste
y_pred = knn.predict(X_test)
```

```python
# Calcular a matriz de confusão
cm = confusion_matrix(y_test, y_pred)
```

```python
# Plotar a matriz de confusão
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Matriz de Confusão")
plt.xlabel("Classe Predita")
plt.ylabel("Classe Real")
plt.show()
```

```python
# Calcular a acurácia das previsões
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia:", accuracy)
```
